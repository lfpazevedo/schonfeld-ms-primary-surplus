## **1. Macroeconomic Context: Forward Rate Spreads and Fiscal Policy in Brazil**

Trading the Brazilian yield curve requires a deep understanding of the unique relationship between **fiscal credibility** and the **term premium**. When analyzing the spread between forward rates like the **1y1y** (the 1-year rate, one year from now) and the **3y3y** (the 3-year rate, three years from now), you are essentially trading the market's view on **long-term structural stability versus cyclical monetary policy**.

### **1.1 Selected Macroeconomic Factor: Primary Surplus Expectations**

While inflation and global interest rates are critical, the most potent factor influencing the slope of the Brazilian curve is **Primary Surplus Expectations**. In Brazil, the primary surplus (government revenue minus non-interest spending) serves as the primary signal for debt sustainability.

#### **Why this factor?**

* **Fiscal Anchor:** It is the most direct proxy for "Fiscal Uncertainty."
* **Sensitivity:** Long-term rates in Brazil are hyper-sensitive to the risk of **"fiscal dominance,"** where monetary policy loses effectiveness because high interest rates actually increase the risk of debt default.

### **1.2 Influence on the Forward Spread (1y1y vs 3y3y)**

The spread between 1y1y and 3y3y represents the **slope of the forward curve**. Under normal conditions, this curve should be upward sloping (positive spread) to compensate for the uncertainty of time.

**How Primary Surplus Expectations shift the spread:**

| Scenario | Mechanism | Spread Impact |
|----------|-----------|---------------|
| **Deteriorating Surplus** | Market demands higher risk buffer for 3y3y | **Steepens** (widens) |
| **Fiscal Improvement** | Risk premium collapses on long end | **Flattens** (narrows) |
| **Fiscal Crisis** | Term premium explosion + BCB hiking | **Volatile/Inversion** |

* **Deteriorating Surplus (Widening Spread):** When expectations for the primary surplus fall, the market demands a higher "risk buffer" for the 3y3y compared to the 1y1y. This causes the curve to steepen. The 1y1y is heavily influenced by the immediate BCB (Central Bank) hiking cycle, but the 3y3y prices in the long-term solvency of the state.
* **Fiscal Improvement (Flattening Spread):** If the government commits to a credible primary surplus, the "fiscal noise" subsides. This reduces the risk premium on the 3y3y more aggressively than on the 1y1y, causing the spread to narrow or the curve to flatten.

### **1.3 The Term-Premium Anomaly: Inversion and Fiscal Risk**

In developed markets, the term premium (the extra compensation for holding long-term debt) is almost always positive. However, in Brazil, we frequently observe **Term-Premium Inversion** or extreme volatility in the long end of the curve.

#### **The Role of Fiscal Uncertainty**

Government fiscal uncertainty acts as a **"tail risk" generator**. In Brazil, when fiscal discipline is questioned, the long end of the curve (3y3y and beyond) doesn't just rise; it can become disconnected from the short-term inflation outlook.

#### **Why Primary Surplus is the Best Macro Factor for Long-Term Trades:**

1. **Direct Impact on the Term Premium:** When the market doubts the primary surplus, the **Term Premium** spikes. Investors fear that the government will eventually have to inflate away the debt or face a credit event.
2. **The Inversion Trigger:** In moments of acute fiscal crisis, the front end (1y1y) may skyrocket as the Central Bank hikes rates to defend the currency. However, if the market believes these hikes will cause a recession that eventually forces rates back down—or if the fiscal situation is so bad that the "terminal rate" is unpredictable—the curve can invert.
3. **Mean Reversion:** Because the Brazilian long-term term premium is so tied to fiscal headlines, "Primary Surplus Expectations" provide a leading indicator for when the curve is "over-punishing" the long end. Trading the 3y3y based on fiscal milestones allows you to capture the compression of that term premium when the government shows even marginal fiscal responsibility.

### **Summary Table: Spread Dynamics**

| Factor Sentiment | 1y1y Reaction | 3y3y Reaction | 1y1y vs 3y3y Spread |
| --- | --- | --- | --- |
| **Primary Surplus Beat** | Stable/Lower | Sharp Decline | **Flattens/Narrows** |
| **Fiscal Target Miss** | Higher (Inflation risk) | Aggressive Spike | **Steepens/Widens** |
| **Fiscal Crisis** | Peak BCB Rates | Term Premium Explosion | **Volatile/Possible Inversion** |

> **💡 Key Insight:** In Brazil, "fiscal noise" often leads to a **"bear steepening"** where long-term rates rise faster than short-term rates, specifically because the primary surplus is seen as the only floor for the term premium.

## **2. Strategy Overview: V5 vs V6**

Both strategies exploit the relationship between **fiscal uncertainty** (measured by cross-sectional dispersion in 4-year primary surplus forecasts) and the **1y1y vs 3y3y curve spread**. However, they differ critically in how they manage **inflation/SELIC regime risk**.

### **2.1 Common Foundation**

| Component | Description |
|-----------|-------------|
| **Primary Signal** | Z-score of `std_4y` (expanding window, no look-ahead) |
| **Secondary Signal** | 2-regime Markov on `diff(std_4y)` for trend confirmation |
| **Trade Structure** | DV01-neutral steepener (long 1y1y, short 3y3y) |
| **Risk Management** | Trailing stop-loss, dynamic DV01, non-linear slippage |

### **2.2 Key Difference: The Inflation Risk Filter**

| Feature | **V5: Advanced Dynamic Size** | **V6: PCA Inflation Filter** |
|---------|-------------------------------|-------------------------------|
| **Regime System** | Single: Fiscal uncertainty only | **Dual:** Fiscal + Inflation/SELIC PCA |
| **Trade Blocking** | ❌ No | ✅ **Yes** (inflation high) |
| **Combined Regimes** | ❌ No | ✅ 4-quadrant classification |
| **Protection Type** | Fiscal only | Fiscal + **Macro filter** |
| **Position Boost** | ❌ No | ✅ +20% for `active_crisis` |
| **Max Position** | 1.0 | 1.2 |

### **2.3 Signal Architecture Comparison**

#### **V5: Two-Layer System**
```
Layer 1: Z-Score of std_4y ──→ Raw Position Size
              ↓
Layer 2: Markov 2-Regime ────→ Scale (0.4x to 1.0x)
```

#### **V6: Three-Layer System with Risk Filter**
```
Layer 0: PCA 3-Regime ───────→ BLOCK if "high" (inflation risk)
              ↓
Layer 1: 4-Quadrant Regime ──→ Classify fiscal environment
              ↓
Layer 2: Markov 2-Regime ────→ Scale position size
```

**The Critical Innovation in V6:** A **3-regime PCA filter** on inflation expectations (IPCA median/std + SELIC median/std). When PC1 indicates "high" inflation expectations, trades are **blocked entirely** — steepening becomes too dangerous when the short end can spike dramatically.

### 6.1 Cumulative P&L

````python
# Extract results
df_v5 = strategy_v5.daily_pnl.copy()
df_v6 = strategy_v6.daily_pnl.copy()

# Calculate cumulative P&L
df_v5['cumulative'] = df_v5['total_pnl'].cumsum()
df_v6['cumulative'] = df_v6['total_pnl'].cumsum()

# Plot
fig, ax = plt.subplots(figsize=(14, 6))

ax.plot(df_v5['date'], df_v5['cumulative'], label='V5 (Dynamic Size)', linewidth=2, alpha=0.8)
ax.plot(df_v6['date'], df_v6['cumulative'], label='V6 (PCA Filter)', linewidth=2, alpha=0.8)

ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Cumulative P&L (bps)', fontsize=12)
ax.set_title('Cumulative P&L Comparison: V5 vs V6', fontsize=14, fontweight='bold')
ax.legend(loc='upper left', fontsize=11)
ax.grid(True, alpha=0.3)

# Add final values
final_v5 = df_v5['cumulative'].iloc[-1]
final_v6 = df_v6['cumulative'].iloc[-1]
ax.axhline(y=final_v5, color='C0', linestyle='--', alpha=0.5)
ax.axhline(y=final_v6, color='C1', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

print(f"\n📊 Final Cumulative P&L:")
print(f"   V5: {final_v5:+.2f} bps")
print(f"   V6: {final_v6:+.2f} bps")
print(f"   Difference: {final_v6 - final_v5:+.2f} bps ({(final_v6/final_v5 - 1)*100:+.1f}%)")
````

### 6.2 Performance Metrics Summary

````python
def calculate_metrics(df):
    """Calculate comprehensive performance metrics."""
    returns = df['total_pnl']
    cumulative = returns.cumsum()
    
    # Basic stats
    total_pnl = returns.sum()
    days = len(df)
    years = days / 252
    
    # Risk metrics
    sharpe = returns.mean() / returns.std() * np.sqrt(252) if returns.std() > 0 else 0
    
    # Drawdown
    peak = cumulative.expanding().max()
    drawdown = cumulative - peak
    max_dd = drawdown.min()
    
    # Win rate
    win_rate = (returns > 0).mean()
    
    return {
        'Total P&L (bps)': total_pnl,
        'P&L/Year (bps)': total_pnl / years,
        'Sharpe Ratio': sharpe,
        'Max Drawdown (bps)': max_dd,
        'Win Rate (%)': win_rate * 100,
        'Trading Days': days,
        'Avg Daily (bps)': returns.mean(),
        'Std Daily (bps)': returns.std(),
    }

metrics_v5 = calculate_metrics(df_v5)
metrics_v6 = calculate_metrics(df_v6)

# Create comparison table
comparison_df = pd.DataFrame({
    'V5 (Dynamic Size)': metrics_v5,
    'V6 (PCA Filter)': metrics_v6,
})
comparison_df['Difference'] = comparison_df['V6 (PCA Filter)'] - comparison_df['V5 (Dynamic Size)']
comparison_df['% Change'] = (comparison_df['V6 (PCA Filter)'] / comparison_df['V5 (Dynamic Size)'] - 1) * 100

print("\n📊 PERFORMANCE METRICS COMPARISON")
print("="*80)
display(comparison_df.round(2))
````

### 6.3 Drawdown Analysis

````python
# Calculate drawdowns
def calc_drawdown(df):
    cumulative = df['total_pnl'].cumsum()
    peak = cumulative.expanding().max()
    return cumulative - peak

df_v5['drawdown'] = calc_drawdown(df_v5)
df_v6['drawdown'] = calc_drawdown(df_v6)

# Plot
fig, ax = plt.subplots(figsize=(14, 5))

ax.fill_between(df_v5['date'], df_v5['drawdown'], 0, alpha=0.4, label='V5 Drawdown')
ax.fill_between(df_v6['date'], df_v6['drawdown'], 0, alpha=0.4, label='V6 Drawdown')

ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Drawdown (bps)', fontsize=12)
ax.set_title('Drawdown Comparison', fontsize=14, fontweight='bold')
ax.legend(loc='lower left', fontsize=11)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n💀 Max Drawdown:")
print(f"   V5: {df_v5['drawdown'].min():.2f} bps")
print(f"   V6: {df_v6['drawdown'].min():.2f} bps")
````

### 7.2 Performance by Fiscal Regime

````python
# Compare performance by fiscal regime
def regime_performance(df, version):
    """Calculate performance by regime."""
    results = []
    for regime in df['regime'].unique():
        if pd.isna(regime):
            continue
        mask = df['regime'] == regime
        subset = df[mask]
        results.append({
            'Version': version,
            'Regime': regime,
            'Days': mask.sum(),
            'Total P&L': subset['total_pnl'].sum(),
            'Avg P&L': subset['total_pnl'].mean(),
            'Win Rate': (subset['total_pnl'] > 0).mean() * 100,
        })
    return pd.DataFrame(results)

regime_v5 = regime_performance(df_v5, 'V5')
regime_v6 = regime_performance(df_v6, 'V6')
regime_comparison = pd.concat([regime_v5, regime_v6], ignore_index=True)

print("\n📈 PERFORMANCE BY FISCAL REGIME")
print("="*80)
display(regime_comparison.round(2))

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Total P&L by regime
pivot_total = regime_comparison.pivot(index='Regime', columns='Version', values='Total P&L')
pivot_total.plot(kind='bar', ax=axes[0])
axes[0].set_title('Total P&L by Fiscal Regime', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Total P&L (bps)')
axes[0].tick_params(axis='x', rotation=45)
axes[0].legend(title='Version')

# Win rate by regime
pivot_win = regime_comparison.pivot(index='Regime', columns='Version', values='Win Rate')
pivot_win.plot(kind='bar', ax=axes[1])
axes[1].set_title('Win Rate by Fiscal Regime', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Win Rate (%)')
axes[1].tick_params(axis='x', rotation=45)
axes[1].legend(title='Version')

plt.tight_layout()
plt.show()
````

### 7.3 V6: Performance by Combined Regime (4-Quadrant)

````python
# Analyze V6's combined regime classification
# First, we need to reconstruct the combined regime classification

def classify_combined_regime(row):
    """Reconstruct V6's 4-quadrant classification."""
    z = row.get('std_4y_zscore', 0)
    ts_vol = row.get('ts_vol_63d_4y_zscore', row.get('ts_vol_63d_4y', np.nan))
    
    if pd.isna(ts_vol):
        return 'unknown'
    
    high_cross_sec = z > 0.5
    high_ts_vol = ts_vol > 0.5
    
    if high_cross_sec and high_ts_vol:
        return 'active_crisis'
    elif high_cross_sec and not high_ts_vol:
        return 'chronic_disagreement'
    elif not high_cross_sec and high_ts_vol:
        return 'consensus_shock'
    else:
        return 'calm'

# Apply classification
df_v6['combined_regime'] = df_v6.apply(classify_combined_regime, axis=1)

# Calculate performance by combined regime
combined_stats = []
for regime in ['active_crisis', 'chronic_disagreement', 'consensus_shock', 'calm', 'unknown']:
    mask = df_v6['combined_regime'] == regime
    if mask.sum() == 0:
        continue
    subset = df_v6[mask]
    combined_stats.append({
        'Regime': regime,
        'Days': mask.sum(),
        'Avg Position': subset['position_size'].mean(),
        'Total P&L': subset['total_pnl'].sum(),
        'Avg P&L': subset['total_pnl'].mean(),
        'Win Rate (%)': (subset['total_pnl'] > 0).mean() * 100,
    })

combined_df = pd.DataFrame(combined_stats)
print("\n📊 V6: PERFORMANCE BY COMBINED REGIME (4-Quadrant)")
print("="*80)
display(combined_df.round(3))

# Visualize
fig, ax = plt.subplots(figsize=(10, 6))

x = range(len(combined_df))
width = 0.35

bars1 = ax.bar([i - width/2 for i in x], combined_df['Total P&L'], width, label='Total P&L', alpha=0.8)
ax2 = ax.twinx()
bars2 = ax2.bar([i + width/2 for i in x], combined_df['Win Rate (%)'], width, label='Win Rate', color='orange', alpha=0.8)

ax.set_xlabel('Combined Regime', fontsize=11)
ax.set_ylabel('Total P&L (bps)', fontsize=11)
ax2.set_ylabel('Win Rate (%)', fontsize=11)
ax.set_title('V6: Performance by Combined Regime', fontsize=13, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(combined_df['Regime'], rotation=45, ha='right')
ax.axhline(y=0, color='black', linestyle='-', linewidth=0.5)

# Legends
ax.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
````

## **9. Risk Attribution**

### 9.1 P&L Decomposition

````python
# Compare P&L attribution
def pnl_attribution(df, version):
    """Calculate P&L attribution."""
    components = ['curve_pnl', 'gamma_pnl', 'carry_pnl', 'cost_pnl']
    totals = {}
    for comp in components:
        if comp in df.columns:
            totals[comp] = df[comp].sum()
    totals['total'] = df['total_pnl'].sum()
    return totals

attr_v5 = pnl_attribution(df_v5, 'V5')
attr_v6 = pnl_attribution(df_v6, 'V6')

# Create comparison
components = ['curve_pnl', 'gamma_pnl', 'carry_pnl', 'cost_pnl']
labels = ['Curve (DV01-neutral)', 'Gamma (convexity)', 'Carry', 'Costs']

v5_values = [attr_v5.get(c, 0) for c in components]
v6_values = [attr_v6.get(c, 0) for c in components]

# Plot
fig, ax = plt.subplots(figsize=(10, 6))

x = range(len(labels))
width = 0.35

bars1 = ax.bar([i - width/2 for i in x], v5_values, width, label='V5', alpha=0.8)
bars2 = ax.bar([i + width/2 for i in x], v6_values, width, label='V6', alpha=0.8)

ax.set_ylabel('Total P&L (bps)', fontsize=11)
ax.set_title('P&L Attribution Comparison', fontsize=13, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=15, ha='right')
ax.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
ax.legend()

# Add value labels
for bar in bars1:
    height = bar.get_height()
    ax.annotate(f'{height:.1f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3 if height >= 0 else -12),
                textcoords="offset points",
                ha='center', va='bottom' if height >= 0 else 'top', fontsize=9)

for bar in bars2:
    height = bar.get_height()
    ax.annotate(f'{height:.1f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3 if height >= 0 else -12),
                textcoords="offset points",
                ha='center', va='bottom' if height >= 0 else 'top', fontsize=9)

plt.tight_layout()
plt.show()

# Table
print("\n📊 P&L ATTRIBUTION")
print("="*70)
attr_df = pd.DataFrame({
    'Component': labels,
    'V5 (bps)': v5_values,
    'V6 (bps)': v6_values,
    'Difference': [v6 - v5 for v5, v6 in zip(v5_values, v6_values)]
})
display(attr_df.round(2))
````

### 9.2 Rolling Sharpe Ratio

````python
# Calculate rolling Sharpe ratio (252-day window)
window = 252

df_v5['rolling_sharpe'] = (df_v5['total_pnl'].rolling(window).mean() / 
                           df_v5['total_pnl'].rolling(window).std()) * np.sqrt(252)
df_v6['rolling_sharpe'] = (df_v6['total_pnl'].rolling(window).mean() / 
                           df_v6['total_pnl'].rolling(window).std()) * np.sqrt(252)

# Plot
fig, ax = plt.subplots(figsize=(14, 5))

ax.plot(df_v5['date'], df_v5['rolling_sharpe'], label=f'V5 ({window}-day)', alpha=0.8)
ax.plot(df_v6['date'], df_v6['rolling_sharpe'], label=f'V6 ({window}-day)', alpha=0.8)

ax.set_xlabel('Date', fontsize=11)
ax.set_ylabel('Sharpe Ratio', fontsize=11)
ax.set_title(f'Rolling {window}-Day Sharpe Ratio Comparison', fontsize=13, fontweight='bold')
ax.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
ax.axhline(y=1, color='green', linestyle='--', alpha=0.5, label='Sharpe = 1')
ax.legend(loc='upper left')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n📈 Rolling Sharpe Statistics (last {window} days):")
print(f"   V5: Mean={df_v5['rolling_sharpe'].mean():.2f}, Last={df_v5['rolling_sharpe'].iloc[-1]:.2f}")
print(f"   V6: Mean={df_v6['rolling_sharpe'].mean():.2f}, Last={df_v6['rolling_sharpe'].iloc[-1]:.2f}")
````
